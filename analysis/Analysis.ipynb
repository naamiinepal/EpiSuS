```julia
using CSV, DataFrames
```

```julia
file = CSV.File("Dataset/nepali_tweets_dataset_vectors_EHnM_analysis_v1.csv")
```

```julia
df = DataFrame(file)
```

```julia
names(df)
```

```julia
using Dates
```

```julia
df.created_at = replace(x->x[1:end-10] * x[end-3:end], df.created_at)
```

```julia
df.created_at = replace(x->x[1:end-13] * x[end-3:end], df.created_at)
```

```julia
df.created_at = Date.(df.created_at,"e u d y")
```

```julia
minimum(df.created_at)
```

```julia
df = sort(df, ["created_at"])
```

```julia
grouped_df = groupby(df, :created_at)
```

```julia
names(df) 
```

```julia
analysed = combine(grouped_df,:covid_stats=>sum, :vaccination=>sum,:covid_politics=>sum,:humour=>sum,:lockdown=>sum,:civic_views=>sum,:life_during_pandemic=>sum, :covid_waves_and_variants=>sum,:misinformation=>sum,:others=>sum)
```

```julia
using Plots
```

```julia
plot(analysed.created_at,[analysed.vaccination_sum,analysed.covid_stats_sum,analysed.covid_politics_sum, analysed.humour_sum, analysed.lockdown_sum,analysed.civic_views_sum,analysed.life_during_pandemic_sum,analysed.misinformation_sum, analysed.others_sum], title="Temporal Analysis", label=["Vaccination","Covid Stats","Politics","Humour","Lockdown","Civic Views","Life During Pandemic","Misinformation","Others"])
```

```julia
# y_2 = rand(23, 10)
y =[analysed.vaccination_sum,analysed.covid_stats_sum,analysed.covid_waves_and_variants_sum,analysed.covid_politics_sum, analysed.humour_sum, analysed.lockdown_sum,analysed.civic_views_sum,analysed.life_during_pandemic_sum,analysed.misinformation_sum, analysed.others_sum]
# size(y[1])
plot(analysed.created_at,y, title="Temporal Analysis")
```
